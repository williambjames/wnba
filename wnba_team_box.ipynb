```python
import pandas as pd

df['game_date'] = pd.to_datetime(df['game_date'])

# Sort by team_id and game_date
df = df.sort_values(by=['team_id', 'game_date']).reset_index(drop=True)

# Calculate days since the last game
df['days_since_last_game'] = df.groupby('team_id')['game_date'].diff().dt.days

# Set to None if the previous game occurred in a different season
df['days_since_last_game'] = df.groupby('team_id').apply(
    lambda group: group.apply(
        lambda row: None if group.index.get_loc(row.name) == 0 or 
                             row['season'] != group.iloc[group.index.get_loc(row.name) - 1]['season'] 
                             else row['days_since_last_game'], 
        axis=1
    )
).reset_index(drop=True)
```

```python
df_func = df.copy()
```

```python
# Select the columns to be used for opponent stats
opp_stats_columns = ['assists', 'blocks', 'defensive_rebounds', 'field_goal_pct', 'field_goals_made', 'field_goals_attempted',
                     'flagrant_fouls', 'fouls', 'free_throw_pct', 'free_throws_made', 'free_throws_attempted', 'largest_lead',
                     'offensive_rebounds', 'steals', 'team_turnovers', 'technical_fouls', 'three_point_field_goal_pct',
                     'three_point_field_goals_made', 'three_point_field_goals_attempted', 'total_rebounds', 'total_technical_fouls',
                     'total_turnovers', 'turnovers']

# Create a new dataframe with opponent stats
opp_stats = df_func[['game_id', 'team_id'] + opp_stats_columns].copy()

# Rename columns to have 'opp_' prefix
opp_stats.columns = ['game_id', 'opponent_team_id'] + ['opp_' + col for col in opp_stats_columns]

# Merge the opponent stats onto the original dataframe
df_func = df_func.merge(opp_stats, how='left', left_on=['game_id', 'opponent_team_id'], right_on=['game_id', 'opponent_team_id'])

df_func
```

```python
import numpy as np

# Define a function to calculate the possession component
def possession_component(fga, fta, oreb, opp_dreb, fgm, tov):
    # Avoid division by zero using np.where
    oreb_rate = np.where(
        (oreb + opp_dreb) > 0,  # Condition
        oreb / (oreb + opp_dreb),  # True: Calculate offensive rebound rate
        0  # False: Set to 0 to avoid division by zero
    )
    # Calculate the possession component
    component = fga + 0.4 * fta - 1.07 * oreb_rate * (fga - fgm) + tov
    return component

# Calculate possessions for the team
df_func['team_possession_component'] = possession_component(
    df_func['field_goals_attempted'],  # FGA
    df_func['free_throws_attempted'],  # FTA
    df_func['offensive_rebounds'],     # OREB
    df_func['opp_defensive_rebounds'], # Opp DREB
    df_func['field_goals_made'],       # FGM
    df_func['turnovers']               # TOV
)

# Calculate possessions for the opponent
df_func['opp_possession_component'] = possession_component(
    df_func['opp_field_goals_attempted'],  # Opp FGA
    df_func['opp_free_throws_attempted'],  # Opp FTA
    df_func['opp_offensive_rebounds'],     # Opp OREB
    df_func['defensive_rebounds'],         # DREB
    df_func['opp_field_goals_made'],       # Opp FGM
    df_func['opp_turnovers']               # Opp TOV
)

# Calculate total possessions
df_func['possessions'] = 0.5 * (
    df_func['team_possession_component'] + df_func['opp_possession_component']
).round(0).astype(int)

# Calculate opponent possessions (same as possessions, since it's a zero-sum game)
df_func['opp_possessions'] = df_func['possessions']
```

```python

# Calculate Offensive Metrics
df_func['ORtg'] = (df_func['team_score'] / df_func['possessions']) * 100
df_func['eFG%'] = (
    (df_func['field_goals_made'] + 0.5 * df_func['three_point_field_goals_made']) / 
    df_func['field_goals_attempted']
) * 100
df_func['TOV%'] = (df_func['turnovers'] / df_func['possessions']) * 100
df_func['FTR'] = df_func['free_throws_attempted'] / df_func['field_goals_attempted']

# Calculate Defensive Metrics
df_func['DRtg'] = (df_func['opponent_team_score'] / df_func['possessions']) * 100
df_func['Opp_eFG%'] = (
    (df_func['opp_field_goals_made'] + 0.5 * df_func['opp_three_point_field_goals_made']) / 
    df_func['opp_field_goals_attempted']
) * 100
df_func['Opp_TOV%'] = (df_func['opp_turnovers'] / df_func['possessions']) * 100

# Calculate Pace
# Assuming each game is 40 minutes long (WNBA standard)
df_func['minutes_played'] = 40
df_func['pace'] = (df_func['possessions'] / df_func['minutes_played']) * 40
```

```python
import pandas as pd

def calculate_aggregated_metrics(df):

    # Sort by team, season, and date
    df = df.sort_values(by=['team_id', 'season', 'game_date'])

    # Define the stats to aggregate
    stats_to_aggregate = [
        'team_score', 'opponent_team_score', 'field_goals_made', 'field_goals_attempted',
        'three_point_field_goals_made', 'free_throws_attempted', 'turnovers',
        'opp_field_goals_made', 'opp_field_goals_attempted', 'opp_three_point_field_goals_made',
        'opp_turnovers', 'possessions'
    ]

    # Calculate rolling sums for the last 10 games (within the same season)
    for stat in stats_to_aggregate:
        df[f'last_10_{stat}'] = df.groupby(['team_id', 'season'])[stat].transform(
            lambda x: x.rolling(window=10, min_periods=1).sum()
        )

    # Calculate cumulative sums for the full season (grouped by season)
    for stat in stats_to_aggregate:
        df[f'season_{stat}'] = df.groupby(['team_id', 'season'])[stat].cumsum()

    # Recalculate metrics for the last 10 games
    df['last_10_ORtg'] = (df['last_10_team_score'] / df['last_10_possessions']) * 100
    df['last_10_DRtg'] = (df['last_10_opponent_team_score'] / df['last_10_possessions']) * 100
    df['last_10_eFG%'] = (
        (df['last_10_field_goals_made'] + 0.5 * df['last_10_three_point_field_goals_made']) / 
        df['last_10_field_goals_attempted']
    ) * 100
    df['last_10_TOV%'] = (df['last_10_turnovers'] / df['last_10_possessions']) * 100
    df['last_10_FTR'] = df['last_10_free_throws_attempted'] / df['last_10_field_goals_attempted']
    df['last_10_Opp_eFG%'] = (
        (df['last_10_opp_field_goals_made'] + 0.5 * df['last_10_opp_three_point_field_goals_made']) / 
        df['last_10_opp_field_goals_attempted']
    ) * 100
    df['last_10_Opp_TOV%'] = (df['last_10_opp_turnovers'] / df['last_10_possessions']) * 100

    # Recalculate metrics for the full season

    df['season_eFG%'] = (
        (df['season_field_goals_made'] + 0.5 * df['season_three_point_field_goals_made']) / 
        df['season_field_goals_attempted']
    ) * 100
    df['season_TOV%'] = (df['season_turnovers'] / df['season_possessions']) * 100
    df['season_FTR'] = df['season_free_throws_attempted'] / df['season_field_goals_attempted']
    df['season_Opp_eFG%'] = (
        (df['season_opp_field_goals_made'] + 0.5 * df['season_opp_three_point_field_goals_made']) / 
        df['season_opp_field_goals_attempted']
    ) * 100
    df['season_Opp_TOV%'] = (df['season_opp_turnovers'] / df['season_possessions']) * 100

    df['season_ORtg'] = (df['season_team_score'] / df['season_possessions']) * 100
    df['season_DRtg'] = (df['season_opponent_team_score'] / df['season_possessions']) * 100
    df['season_NetRtg'] = df['season_ORtg'] - df['season_DRtg']

    return df
```
